### Perceptron that classifies "shapes" into predators or preys.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ref_prey = np.zeros(21, dtype=np.float64)
ref_prey[[9, 10]] = 1
ref_pred = np.zeros(21, dtype=np.float64)
ref_pred[3:17] = 1
plt.plot(ref_prey)
plt.plot(ref_pred)
```

```python
def gen_data(ref, std=0.3, num=100, plot=True):
    d = np.abs(np.random.normal(ref, std, size=(num, 21)))
    d_min = d.min(axis=1)
    d_max = d.max(axis=1)
    d[:, -1] = d_max
    d = (d - d_min[:, None]) / (d_max - d_min)[:, None]
    if plot:
        for i in range(num): plt.plot(d[i, :-1])
    return d
```

```python
prey = gen_data(ref_prey, num=30)
```

```python
pred = gen_data(ref_pred, num=30)
```

```python
data = np.append(prey, pred, axis=0)
labels = np.ones(data.shape[0])
labels[data.shape[0]//2:] = -1
indices = np.arange(start=0, stop=data.shape[0])
np.random.shuffle(indices)
data = data[indices]
labels = labels[indices]
```

```python
w = np.random.uniform(-1, 1, 21)

for i in range(data.shape[0]): 
    for j in range(data.shape[0]):
        o = w @ data[j]
        o = np.tanh(o)
        
        d_err = o - labels[j]
        
        w -= d_err * (1 - o**2) * data[j]
```

```python
def test(test, true_label, plot=False):
    err = 0
    for i in range(test.shape[0]):
        o = w @ test[i]
        o = np.tanh(o)
        err += (o - true_label)**2
        if plot:
            plt.scatter(o, 0)
            plt.xlim([-1, 1])
    err /= test.shape[0]
    print('MSE: ', err)
```

```python
prey_test = gen_data(ref_prey, plot=False)
test(prey_test, 1, True)
```

```python
pred_test = gen_data(ref_pred, plot=False)
test(pred_test, -1, True)
```
